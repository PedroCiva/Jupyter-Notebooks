For writing into to cvs file

```python
import csv
import openpyxl
import os
```

```python
rola = 0
for excelFile in os.listdir('spreadsheets'):
    print(excelFile)
    # Get workbook
    wb = openpyxl.load_workbook(os.path.join('spreadsheets',excelFile))
    # Loop through all sheets inside of the excel file
    for sheetName in wb.get_sheet_names():
        sheet = wb.get_sheet_by_name(sheetName)
        # Create new filename with .csv extension
        newFileName = excelFile.replace('.xlsx','.csv')

        # Create new csv file and csv writer object
        output_file = open(os.path.join('converted spreadsheets', newFileName), 'w', newline='')
        output_writer = csv.writer(output_file)

        # Write rows in the new csv file
        for row in range(1, sheet.max_row + 1):
            # append each cell to this list
            rowData = []
            # Loop through each cell in the row
            for col in range(1, sheet.max_column + 1):
                rowData.append(sheet.cell(row=row, column=col).value)
            output_writer.writerow(rowData)
        output_file.close()
```
